```python
import openpyxl as xl
import pandas as pd
import re
```

```python
file = xl.load_workbook("w38.xlsx")
```

```python
sheet = file["Sheet1"]
```

```python
dates_dict = {}
for cell in sheet[3]:
    if re.match(r'\d+\.\d+\.\d+', str(cell.value)):
        dates_dict[cell.value] = {}
```

```python
dates_array = []
for date in dates_dict:
    dates_array.append(date)
print(dates_array)
```

```python
week = 0
for cell in sheet[2]:
    if re.match(r'\d+', str(cell.value)):
        week = cell.value
print(week)
```

```python
year = 0
for cell in sheet[1]:
    if re.match(r'\d+', str(cell.value)):
        year = cell.value
print(year)
```

```python
df1 = pd.DataFrame(sheet.values)
df1 = df1.drop([0,1,2])
display(df1)
```

```python
# This calls the first row for the header
new_header = df1.iloc[0] 
# take the rest of your data minus the header row
df1 = df1[1:] 
# set the header row as the df1 header
df1.columns = new_header 
# Lets see the 5 first rows of the new dataset
df1.head()
```

```python
display(df1.columns)
```

```python
new_header = []
i = 0
for name in df1.columns:
    if name == "Остаток":
        new_header.append(f"{dates_array[i]}")
        i += 1
        continue
    new_header.append(name)
print(new_header)
df1.columns = new_header
df1
```

```python
df1
```

```python
df1 = df1.drop(["Страховой запас", "Блокировка"], axis=1)
```

```python
for index, row in df1.iterrows():
    stock = row[dates_array[0]]
    dArrIdx = 1
    if stock == "-":
        while stock == "-":
            if row[dates_array[dArrIdx]] != "-":
                row[dates_array[0]] = row[dates_array[dArrIdx]]
                stock = row[dates_array[dArrIdx]]
            dArrIdx += 1
```

```python
columns_to_delete = dates_array[1:]
print(columns_to_delete)
```

```python
df1 = df1.drop(columns_to_delete, axis=1)
df1
```

```python
df1.insert(0, "Неделя", [week]*(len(df1.index)))
df1.insert(0, "Год", [year]*(len(df1.index)))
df1
```

```python
df1 = df1.rename(columns={str(dates_array[0]): "Остаток"})
df1
```

```python
fileName = f"X5_Stock_{week}.xlsx"
df1.to_excel(fileName, index=0)
```
